## **Data Visualization**

```python
Data['Date of Observation'] = pd.to_datetime(Data['Date of Observation'], dayfirst=True, errors='coerce')
Data['Year'] = Data['Date of Observation'].dt.year
```

```python
plt.figure(figsize=(8, 6))
for species, sub in Data.groupby('Common Name'):
    plt.scatter(sub['Observed Length (m)'], sub['Observed Weight (kg)'], label=species, s=50)
plt.xlabel('Observed Length (m)')
plt.ylabel('Observed Weight (kg)')
plt.title("Crocodile Length vs Weight by Species")
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(title='Species', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
notes_text = " ".join(
    Data["Notes"].dropna().astype(str).tolist()
)

custom_stops = STOPWORDS.union({
    "mr"
})

wc_notes = WordCloud(
    width=1200,
    height=600,
    background_color="white",
    stopwords=custom_stops,
    collocations=True  # keep common phrases from notes
).generate(notes_text)

plt.figure(figsize=(8,6))
plt.imshow(wc_notes, interpolation="bilinear")
plt.axis("off")
plt.title("Notes Word Cloud")
plt.tight_layout()
plt.show()
```

## **Observers gender**

```python
Data["FirstName"] = Data["Observer Name"].str.split().str[0]

d = gender.Detector(case_sensitive=False)
Data["GenderGuess"] = Data["FirstName"].apply(d.get_gender)

# Map categories to Male/Female buckets
mapping = {
    "male": "Male",
    "mostly_male": "Male",
    "female": "Female",
    "mostly_female": "Female",
    "andy": "Unknown",   # androgynous
    "unknown": "Unknown"
}
Data["Gender"] = Data["GenderGuess"].map(mapping)

plt.figure(figsize=(4, 4))
gender_counts = Data["Gender"].value_counts()
gender_counts.plot(kind="bar", color=["pink", "blue", "gray"])
plt.title("Observer Gender Distribution")
plt.ylabel("Count")
plt.show()
```
